```python
import os
import random
import shutil
from sklearn.model_selection import train_test_split
```

```python
image_folder = "Data/images"
```

```python
train_ratio = 0.85
test_ratio = 0.15

random_seed = 42
```

```python
image_files = os.listdir(image_folder)
```

```python
import os

image_file_path = 'Data/images/train/frame3_jpg.rf.77a50289f02cc37a3be3e2faaf0d7885.jpg'
image_file_name = os.path.basename(image_file_path)
image_name = os.path.splitext(image_file_name)[0]

print("Image file name:", image_file_name)
print("Image name:", image_name)
```

```python

train_files, val_files = train_test_split(image_files, train_size=train_ratio, test_size=test_ratio, random_state=random_seed)
```

```python
import re
numbers = re.compile(r'(\d+)')
def numericalSort(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts
```

```python
train_files=sorted(train_files,key=numericalSort)
val_files=sorted(val_files,key=numericalSort)
```

```python
train_files
```

```python


train_folder = os.path.join(image_folder, 'train')
val_folder = os.path.join(image_folder, 'val')
```

```python
os.makedirs(train_folder, exist_ok=True)
os.makedirs(val_folder, exist_ok=True)
```

```python

for file_name in train_files:
    source_path = os.path.join(image_folder, file_name)
    destination_path = os.path.join(train_folder, file_name)
    shutil.move(source_path, destination_path)

for file_name in val_files:
    source_path = os.path.join(image_folder, file_name)
    destination_path = os.path.join(val_folder, file_name)
    shutil.move(source_path, destination_path)
```

```python
train_list=[]
for train in train_files:
    image_file_path = train
    image_file_name = os.path.basename(image_file_path)
    image_name = os.path.splitext(image_file_name)[0]
    train_list.append(image_name)

```
